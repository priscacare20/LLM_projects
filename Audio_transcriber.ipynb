### convet audio to text using openAI's whisper model.

```python
# Specifies the pre-trained Whisper Medium model hosted on Hugging Face. This model is capable of converting speech into text.
audio_model = "openai/whisper-medium"
speech_model = AutoModelForSpeechSeq2Seq.from_pretrained(audio_model, torch_dtype=torch.float16, low_cpu_mem_usage=True, use_safetensors=True)

# Loads the Whisper model optimized for speech-to-text tasks.
# torch_dtype=torch.float16: Uses 16-bit precision to speed up computation and reduce memory usage (great for GPUs).
# low_cpu_mem_usage=True: Reduces RAM usage during loading.
# use_safetensors=True: Loads model weights from safer, faster .safetensors format if available.

speech_model.to('cuda') #  move the model to GPU

# process the audio using autoprocessor. This uses feature extractor to extract the text and a tokenize to process the text into a format useful to the model
processor = AutoProcessor.from_pretrained(audio_model)

pipe = pipeline(
    "automatic-speech-recognition",
    model=speech_model,
    tokenizer=processor.tokenizer,
    feature_extractor=processor.feature_extractor,
    torch_dtype=torch.float16,
    device='cuda',
)
```

```python
# Use the Whisper OpenAI model to convert the Audio to Text
# return_timestamps = True, is used when the audio is longer than 3000 mel, language =en is used to translate the transcripttion to English
result = pipe(audio_filename, return_timestamps=True)

transcription = result["text"]
```

### summarize audio content using llama

```python
# make a minute prompt message for the llm

system_message = "You are an assistant that produces minutes of meetings from transcripts, with summary, key discussion points, takeaways and action items with owners, in markdown."
user_prompt = f"Below is an extract transcript of a Denver council meeting. Please write minutes in markdown, including a summary with attendees, location and date; discussion points; takeaways; and action items with owners.\n{transcription}"

messages = [
    {"role": "system", "content": system_message},
    {"role": "user", "content": user_prompt}
  ]
```

```python
# create a configuration object to tell the transformer how to load the model using quantization. Quantization is a technicque
# that reduces model size and increases inference speed by using lower precision numbers.
# loads the model in 4-bit precision instead of the default 16-bit.
# bnb_4bit_use_double_quant= True, enable double quantization. firs the models weight is quantized to 8-bit and then inot 4-bit to preserve accuracy.
# nf4 --> normalized float 4 quantization method. it retains model accuracy more than older methods like fp4

quant_config = BitsAndBytesConfig(
    load_in_4bit=True,
    bnb_4bit_use_double_quant=True,
    bnb_4bit_compute_dtype=torch.bfloat16,
    bnb_4bit_quant_type="nf4"
)
```

```python
# load llama tokenizer and set the pad token to be equal to end-of-sequence token
llama = "meta-llama/Meta-Llama-3.1-8B-Instruct"
tokenizer = AutoTokenizer.from_pretrained(llama)
tokenizer.pad_token = tokenizer.eos_token

# tokenize the prompts and stream the output for real time generation
inputs = tokenizer.apply_chat_template(messages, return_tensors="pt").to("cuda")
streamer = TextStreamer(tokenizer)
model = AutoModelForCausalLM.from_pretrained(llama, device_map="auto", quantization_config=quant_config)
outputs = model.generate(inputs, max_new_tokens=2000, streamer=streamer)
```

```python
# decode the output
response = tokenizer.decode(outputs[0])
display(Markdown(response))
```
